**TODO 6: Corregir la fecha de nacimiento en funcion de la edad para todos los registros que lo requieran**

```python
homicides_df[["AÑO DE NACIMIENTO", "EDAD"]].head(10)
```

```python
currentDateTime = datetime.now()
date = currentDateTime.date()
year = date.strftime("%Y")
print ('Año actual: ' + year)

for i in homicides_df.index:
  edad = homicides_df['EDAD'][i]
  birthYear = int(year) - int(edad)
  homicides_df['AÑO DE NACIMIENTO'][i] = birthYear
```

*En la celda anterior se puede evidenciar que los registros aún conservan la edad que tenian originalmente, pero que al comparar con la celda ubicada dos celdas atrás el año de nacimiento es diferente. El año de nacimiento se calculó en función de la edad y tomando como referencia el año actual (que se obtiene mediante la fecha del sistema)*
